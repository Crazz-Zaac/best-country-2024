```python
# import statistics as stat
from scipy import stats
```

```python
from scipy.stats import wilcoxon
res = wilcoxon(clean_df['Hdi2020'], clean_df['Hdi2021'], alternative='greater')
res.statistic, res.pvalue
```

```python
res
```

```python
stats.shapiro(clean_df['Hdi2020'])
```

```python
stats.ttest_rel(clean_df['Hdi2020'], clean_df['Hdi2021'])
```

```python
test_result = stats.ttest_rel(clean_df['Hdi2020'], clean_df['Hdi2021'], alternative = "less")

print("t-value:", round(test_result.statistic, 5))
print("p-value:", round(test_result.pvalue, 5))
```
